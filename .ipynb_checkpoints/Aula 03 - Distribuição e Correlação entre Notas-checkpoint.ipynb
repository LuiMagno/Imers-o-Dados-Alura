```python
# Desafio 03 - Criar uma função para plotar o boxplot do seaborn
```

```python
def plotar_grafico(eixoX,eixoY,dadosTemp,titulo, ordem):
    plt.figure(figsize=(10,8))
    sns.boxplot(x= eixoX, y = eixoY, data = dadosTemp, order = ordem)
    plt.title(titulo)
```

```python
plotar_grafico('Q006', 'NU_NOTA_TOTAL', dados, 'Gráfico Notas de Matemática em relação a renda', renda_ordenada)
```
