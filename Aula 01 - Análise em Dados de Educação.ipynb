```python
# Desafio 03: Adicionar título no gráfico
# Utilizar a biblioteca matplot que tem na funçao 'plot()' e com o parametro 'title' mudar o título do gráfico.
```

```python
import matplotlib.pyplot as plt
dados['NU_NOTA_REDACAO'].plot.hist(bins=20, figsize=(8, 6), title="Grafico de distribuição das notas de redação", grid=True)
```

```python
# Desafio 04: Plotar os Histogramas das idades dos do treineiro e não treineiros.
# Utilizando os plots e as query para plotar os gráficos de idades em relação aos treineiros e não treineiros
```

```python
dados_nao_treineiro = dados.query('IN_TREINEIRO == 0')
dados_nao_treineiro['NU_IDADE'].plot.hist(grid = True, bins = 20, title = 'Não Treineiros por Idade')
```

```python
dados_treineiro = dados.query('IN_TREINEIRO == 1')
dados_treineiro['NU_IDADE'].plot.hist(grid = True, bins = 20, title = 'Treineiros por Idade')
```
